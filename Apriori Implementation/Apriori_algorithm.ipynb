##Importing libraries

```python
import numpy as np

import pandas as pd

import matplotlib.pyplot as plt
```

##Importing dataset

```python
# Load the dataset
dataset = pd.read_csv('bigBasket.csv', header=None)

transactions = []

# Assuming the dataset has fewer columns than 20, we need to loop only up to the number of columns in the dataset
num_columns = dataset.shape[1]  # Get the number of columns in the dataset

for i in range(0, 7219):
  transactions.append([str(dataset.values[i, j]) for j in range(0, num_columns)])
```

```python
print(dataset)
```

```python
print(transactions)
```

#Apriori Training on Dataset

```python
from apyori import apriori

# Apply the apriori algorithm
basket_intelligence = apriori(
    transactions=transactions,
    min_support=0.002,
    min_confidence=0.2,
    min_lift=3,
    min_length=2,
    max_length=2
)
```

#Visualizing

##Raw result

```python
results = list(basket_intelligence)
results
```

##Proper display

```python
def inspect(results):
    product1 = [tuple(result[2][0][0])[0] for result in results]
    product2 = [tuple(result[2][0][1])[0] for result in results]
    supports = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts = [result[2][0][3] for result in results]
    return list(zip(product1, product2, supports, confidences, lifts))

DataFrame_intelligence = pd.DataFrame(inspect(results), columns=['product1', 'product2', 'Support', 'Confidence', 'Lift'])
```
